```python
import serialComunication
import time
```

```python

arduino = serialComunication.com_serial(dir_com="/dev/ttyUSB0")
```

Primer elemento necesario para la comunicacion serial es el iniciar el objeto con el que se quiere estableser la comunicación con sus parametros correspondientes. En este caso el "/dev/ttyUSB0" en linux (o COMx para windows) este puede variar tanto la notación como la numeracion según el sistema operativo como tambien del puerto desde el que se conecta. Es por lo anterior se que recomienda primero identificar el puerto a usar y no cambiarlo hasta que se vuelva a corregir la inicializacion del objeto de comunicacion serial. 

```python
print(arduino._com_serial__channel)
arduino.init_com()
```

Se comprueba que no existe un objeto com_serial en memoria del objeto inicializado como arduino, posteriormente se se establese la comunicacion con todas sus caracteristicas: id,apertura del canal y su configuracion de comunicacion destacando el puerto(port), el baudrate y el timeout. Si se desea se puede estableser otra configuracion incial en el codigo de la clase o por medio del metodo init de serialComnication.com_serial().

```python
time.sleep(2)
arduino.send_comand(0,'c')
```

Para el envio de informacion existen 3 tipos de envio de informacion: Modo 'a' es el modo default en el que se ingresa un string que es formateado y enviado directamente al dispositivo; Modo 'b' este se debe senalar al momento de usarlo, en este modo es donde se puede interactuar por medio de la consola para poder enviar el string ingresado manualmente, ideal para pruebas de diferentes metodos y comandos; Modo 'c' se tienen que senalar, es en este modo en el que se puede recurrir a una tabla de convercion para enviar un comando por medio de numeros
